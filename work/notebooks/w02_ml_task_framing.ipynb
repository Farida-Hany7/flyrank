## 1. My lane as an ML task (type)

*Classification, clustering, ranking, or scoring — which one, and why?*

## My lane as an ML task

My lane is **Content Decline Prioritization**.

The goal is to help editors identify which content pages should be reviewed first.

This is a **classification** problem because the model predicts whether a page should be prioritized for review or not.

## 2. Target or proxy

*What would you predict? Where does that label come from — observed outcome or a defined rule?*

## Target or proxy

The dataset does not contain a direct label for "needs review," so I will create a proxy target.

Target:

needs_review

Values:

- 1 = content should be prioritized for editorial review
- 0 = content does not currently need review

The proxy target can be created using observed signals such as a downward trend and a long time since the last content update.

## 3. Success metric

*One metric you can defend. What number means 'good'?*

## Success metric

The main success metric is Precision.

Editors have limited time, so the model should recommend pages that genuinely need review.

Recall is also useful because missing important declining pages reduces the value of the recommendations.

```python
cols = [
    "content_id",
    "client_id",
    "trend_direction",
    "trend_pct",
    "days_since_last_update",
    "search_volume",
    "clicks_90d",
    "impressions_90d"
]

df[cols].head()
```

Each row represents one content page.

Each page contains information about search performance, engagement, freshness, and traffic trends. The model will use these features to estimate whether the page should be prioritized for review.

```python
df["needs_review"] = (
    (df["trend_direction"] == "down") &
    (df["days_since_last_update"] > 30)
).astype(int)

df[["trend_direction",
    "days_since_last_update",
    "needs_review"]].head()
```

## 5. Why ML beats a fixed rule here

*What makes the pattern too messy for an if-statement?*

## Why ML beats a fixed rule

A simple rule, such as reviewing every page older than 30 days, ignores many useful signals.

Machine learning can combine several features including:

- trend direction
- trend percentage
- search volume
- click-through rate
- impressions
- engagement rate
- content age

to identify more complex patterns than a single manually written rule.

This helps editors prioritize pages more effectively and makes better use of their limited review time.
